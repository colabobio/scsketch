```python
import scanpy as sc
import urllib.request
from pathlib import Path

URL = "https://datasets.cellxgene.cziscience.com/c3d1a5e6-780b-4fe9-a39b-1864f927e87b.h5ad"
DATA_PATH = Path("../../../1M_20260121.h5ad")
DATA_PATH.parent.mkdir(parents=True, exist_ok=True)

if not DATA_PATH.exists():
    # print("File doesn't exist")
    print("Downloading CZI CMV 1M demo...")
    urllib.request.urlretrieve(URL, DATA_PATH)
    print("Download complete.")



adata_ui = sc.read_h5ad(DATA_PATH)
print(f"Loaded {adata_ui.shape[0]:,} cells × {adata_ui.shape[1]:,} genes")
```
